```python
import pandas as pd
import os

# Read the CSV file from current directory
csv_file = 'musae_PTBR_edges.csv'

# Read the first CSV file found
df = pd.read_csv(csv_file)

# Check if 'from' and 'to' columns exist
if 'from' in df.columns and 'to' in df.columns:
    # Create output text file name
    output_file = os.path.splitext(csv_file)[0] + '.txt'
    
    # Write to txt file with columns separated by a blank line
    with open(output_file, 'w', encoding='utf-8') as f:
        for _, row in df.iterrows():
            f.write(f"{row['from']} {row['to']}\n")
            
else:
    print("CSV file does not contain 'from' and 'to' columns")
```
